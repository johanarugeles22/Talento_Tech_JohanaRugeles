```python
#Lista detallada de juegos en Spain
query= """SELECT
            m.id AS match_id,
            c.name AS country_name,
            l.name AS league_name,
            m.season,
            m.stage,
            m.date,
            th.team_long_name AS home_team,
            ta.team_long_name AS away_team,
            m.home_team_goal,
            m.away_team_goal
        FROM
            Match m
        JOIN
            Country c ON m.country_id = c.id
        JOIN
            League l ON m.league_id = l.id
        JOIN
            Team th ON m.home_team_api_id = th.team_api_id
        JOIN
            Team ta ON m.away_team_api_id = ta.team_api_id
        WHERE
            c.name = 'Spain'
        ORDER BY
            m.date ASC
        LIMIT 10;"""

detailed_matches = pd.read_sql(query, conn)
detailed_matches
```

Sentencia HAVING

```python
#Análisis de goles por temporada en las ligas de España, Francia e Inglaterra
query = """ SELECT season,
            Country.name                          AS Country,
            League.name                           AS League,
            ROUND(AVG(Match.home_team_goal),2)    AS avg_local_goals,    --Promedio goles locales
            ROUND(AVG(Match.away_team_goal),2)    AS avg_visitor_goals,  --Promedio goles visitantes
            SUM(home_team_goal + away_team_goal)  AS total_goals
        FROM Match
            JOIN Country ON Country.id = Match.country_id
            JOIN League ON League.id = Match.league_id
        WHERE Country.name IN ('Spain', 'France', 'England')
        GROUP BY season, League
        HAVING avg_local_goals BETWEEN 1.4 AND 1.6 AND avg_visitor_goals BETWEEN 1.2 AND 1.6
        ORDER BY season DESC;"""

goals_by_season = pd.read_sql(query, conn)
goals_by_season
```

```python
#Para las ligas de los paises Spain, Germany, France, Italy y England que tengan un promedio de goles por partido mayor a 2.5

query= """ SELECT
            c.name AS Country,
            l.name AS League,
            COUNT(DISTINCT m.stage) AS Stages,
            m.season AS Season,
            COUNT(DISTINCT m.home_team_api_id) AS number_of_teams,
            ROUND(AVG(m.home_team_goal), 2) AS avg_local_goals,
            ROUND(AVG(m.away_team_goal), 2) AS avg_visitor_goals,
            ROUND(AVG(m.home_team_goal + m.away_team_goal), 2) AS avg_goals,
            ROUND(AVG(m.home_team_goal - m.away_team_goal), 2) AS avg_goal_dif,
            SUM(m.home_team_goal + m.away_team_goal) AS total_goals
        FROM
            Match m
        JOIN
            Country c ON m.country_id = c.id
        JOIN
            League l ON m.league_id = l.id
        WHERE
            c.name IN ('Spain', 'Germany', 'France', 'Italy', 'England')
        GROUP BY
            c.name, l.name, m.season
        HAVING
            AVG(m.home_team_goal + m.away_team_goal) > 2.5
        ORDER BY
            c.name, l.name, m.season; """

leages_by_season = pd.read_sql(query, conn)
leages_by_season
```

```python
#Análisis sobre los equipos que más goles realizaron jugando como locales
# en las ligas de Spain, Germany y England, hayan anotado más de 50 goles
query= """SELECT
    Country.name AS Country,
    League.name AS League,
    Team.team_long_name AS Team,
    COUNT(DISTINCT Match.stage) AS Stages,
    Match.season AS Season,
    SUM(Match.home_team_goal) AS sum_goals,
    ROUND(AVG(Match.home_team_goal), 2) AS avg_goals
FROM
    Match
JOIN Country ON Country.id = Match.country_id
JOIN League ON League.id = Match.league_id
JOIN Team ON Team.team_api_id = Match.home_team_api_id
WHERE
 Country.name IN ('Spain', 'Germany', 'England')
GROUP BY
    Country.name, League.name, Team.team_long_name, Match.season
HAVING
    SUM(Match.home_team_goal) > 50
ORDER BY
    Match.season DESC, sum_goals DESC;"""


teams_by_season = pd.read_sql(query, conn)
teams_by_season
```

Sentencia CASE WHEN en SELECT

```python
#indique los partidos jugados por el FC Barcelona, su rival, la fecha de juego y si perdió, ganó o empato
query= """SELECT DATE(Match.date) as MatchDate,
                Home.team_long_name AS  HomeTeam,
                home_team_goal AS Home_goals,
                Visitor.team_long_name AS VisitorTeam,
                away_team_goal Visitor_goals,
                CASE
                WHEN Home.team_api_id = 8634 AND home_team_goal < away_team_goal THEN 'Barca lost :('
                    WHEN Home.team_api_id = 8634 AND home_team_goal > away_team_goal THEN 'Barca won'
                    WHEN Visitor.team_api_id = 8634 AND home_team_goal < away_team_goal THEN 'Barca won'
                    WHEN Visitor.team_api_id = 8634 AND home_team_goal > away_team_goal THEN 'Barca lost :('
                    ELSE 'Tie'
                END AS Outcome
        FROM Match
        LEFT JOIN Team AS Home on Home.team_api_id = Match.home_team_api_id
        LEFT JOIN Team AS Visitor on Visitor.team_api_id = Match.away_team_api_id
        WHERE (Visitor.team_api_id = 8634 OR Home.team_api_id = 8634) AND season = "2012/2013"
        Order by MatchDate DESC
        LIMIT 15;"""

Barca_results = pd.read_sql(query, conn)
Barca_results
```

```python
#indique los partidos jugados por el FC Barcelona como local contra el Real Madrid CF, la fecha de juego y si perdió, ganó o empató
#12.2
query="""SELECT
    DATE(M.date) AS MatchDate,
    HomeTeam.team_long_name AS HomeTeam,
    VisitorTeam.team_long_name AS VisitorTeam,
    M.home_team_goal AS HomeGoals,
    M.away_team_goal AS AwayGoals,
    CASE
        WHEN M.home_team_goal > M.away_team_goal THEN 'Ganó'
        WHEN M.home_team_goal < M.away_team_goal THEN 'Perdió'
        ELSE 'Empató'
    END AS Outcome
FROM
    Match AS M
JOIN
    Team AS HomeTeam ON M.home_team_api_id = HomeTeam.team_api_id
JOIN
    Team AS VisitorTeam ON M.away_team_api_id = VisitorTeam.team_api_id
WHERE
    (M.home_team_api_id = 8634 AND M.away_team_api_id = 8633)
ORDER BY
    MatchDate;"""

BarcaMadrid_results = pd.read_sql(query, conn)
BarcaMadrid_results
```

SUB QUERIES

```python
#Rating de Cristiano Ronaldo y Lionel Messi versus los demás jugadores
query = """SELECT CASE
                   WHEN player_name = 'Cristiano Ronaldo' THEN 'Cristiano Ronaldo'
                   WHEN player_name = 'Lionel Messi' THEN 'Lionel Messi'
                   ELSE 'Others'
                   END AS Player,
           PA_Grouped.avg_overall_rating AS avg_overall_rating,
           PA_Grouped.avg_potential AS avg_potential
           FROM Player
           LEFT JOIN
           (SELECT Player_Attributes.player_api_id,
                         ROUND(avg(Player_Attributes.overall_rating),2) AS avg_overall_rating,
                         ROUND(avg(Player_Attributes.potential),2) AS avg_potential
                     FROM Player_Attributes
                     GROUP BY Player_Attributes.player_api_id)
                    AS PA_Grouped
                    ON Player.player_api_id = PA_Grouped.player_api_id
           GROUP BY Player
           ORDER BY Player;"""

players_rating = pd.read_sql(query, conn)
players_rating
```

```python
#Comparación de tipo de juego entre el FC Barcelona y Real Madrid FC - SUB QUERY
#13.2
query="""SELECT
    TA.date AS RatingDate,
    TA.team_api_id AS TeamID,
    T.team_long_name AS Team,
    -- Clasificación de buildUpPlaySpeed como PlaySpeed
    CASE
        WHEN TA.buildUpPlaySpeed >= 70 THEN 'Fast'
        WHEN TA.buildUpPlaySpeed <= 30 THEN 'Slow'
        ELSE 'Balanced'
    END AS PlaySpeed,
    -- Clasificación de buildUpPlayPassing como PlayPassing
    CASE
        WHEN TA.buildUpPlayPassing >= 70 THEN 'Long'
        WHEN TA.buildUpPlayPassing <= 30 THEN 'Short'
        ELSE 'Mixed'
    END AS PlayPassing,
    -- Clasificación de chanceCreationShooting como CreationShooting
    CASE
        WHEN TA.chanceCreationShooting >= 68 THEN 'Lots'
        WHEN TA.chanceCreationShooting <= 30 THEN 'Little'
        ELSE 'Normal'
    END AS CreationShooting
FROM
    Team_Attributes AS TA
JOIN
    Team AS T ON TA.team_api_id = T.team_api_id
GROUP BY
    RatingDate, TeamID, Team
ORDER BY
    RatingDate, TeamID;
"""

BarcaMadrid_rating = pd.read_sql(query, conn)
BarcaMadrid_rating
```

```python
#Comparación de tipo de juego entre el FC Barcelona y Real Madrid FC - WHERE
#13.3

query= """SELECT
    TA.date AS RatingDate,
    TA.team_api_id AS TeamID,
    T.team_long_name AS Team,
    -- Clasificación de buildUpPlaySpeed como PlaySpeed
    CASE
        WHEN TA.buildUpPlaySpeed >= 70 THEN 'Fast'
        WHEN TA.buildUpPlaySpeed <= 30 THEN 'Slow'
        ELSE 'Balanced'
    END AS PlaySpeed,
    -- Clasificación de buildUpPlayPassing como PlayPassing
    CASE
        WHEN TA.buildUpPlayPassing >= 70 THEN 'Long'
        WHEN TA.buildUpPlayPassing <= 30 THEN 'Short'
        ELSE 'Mixed'
    END AS PlayPassing,
    -- Clasificación de chanceCreationShooting como CreationShooting
    CASE
        WHEN TA.chanceCreationShooting >= 68 THEN 'Lots'
        WHEN TA.chanceCreationShooting <= 30 THEN 'Little'
        ELSE 'Normal'
    END AS CreationShooting
FROM
    Team_Attributes AS TA
JOIN
    Team AS T ON TA.team_api_id = T.team_api_id
WHERE
    -- Clasificación de buildUpPlaySpeed
    (TA.buildUpPlaySpeed >= 70 OR TA.buildUpPlaySpeed <= 30)
    AND
    -- Clasificación de buildUpPlayPassing
    (TA.buildUpPlayPassing >= 70 OR TA.buildUpPlayPassing <= 30)
    AND
    -- Clasificación de chanceCreationShooting
    (TA.chanceCreationShooting >= 68 OR TA.chanceCreationShooting <= 30)
ORDER BY
    RatingDate, TeamID;"""

subquery2 = pd.read_sql(query, conn)
subquery2
```
